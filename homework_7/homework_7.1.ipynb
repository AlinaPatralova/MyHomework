## Задание 1. Получение данных из публичного API

Выберите публичный API. Например, JSONPlaceholder.

Напишите скрипт, который:
1. Отправляет GET-запрос к /posts,
2. Извлекает и выводит заголовки и тела первых 5 постов.

```python
import requests

base_url = "https://jsonplaceholder.typicode.com"

def main():
    url = f"{base_url}/posts"
    resp = requests.get(url, timeout=10)
    resp.raise_for_status()
    posts = resp.json() 

    for i, post in enumerate(posts[:5], start=1):
        print(f"{i}) {post['title']}\n\n{post['body']}\n\n{'-'*60}")

if __name__ == "__main__":
    main()
    
```
