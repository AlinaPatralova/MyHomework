## Задание 4. Обработка ошибок и работа с данными

1. Расширьте предыдущий код для обработки разных кодов состояния, например: 400, 404.
2. Добавьте вывод сообщения об ошибке в зависимости от полученного кода состояния.

```python
import json
import requests

api_url = "https://jsonplaceholder.typicode.com/posts"

new_post = {
    "title": "Привет, POST!",
    "body": "Тестовый контент"
}

try:
    r = requests.post(api_url, json=new_post, timeout=10)
    r.raise_for_status()  # если код >=400 - улетим в except HTTPError
    data = r.json()

    post_id = data.get("id")
    print(f"ID созданного поста: {post_id}")
    print("Содержимое созданного поста:")
    print(json.dumps(data, ensure_ascii=False))

except requests.exceptions.HTTPError as e:
    status = e.response.status_code if e.response is not None else None
    if status == 400:
        print("Ошибка: Неверный запрос (400), проверьте тело/параметры")
    elif status == 404:
        print("Ошибка: Ресурс не найден (404), проверьте URL")
    elif status is not None and 500 <= status < 600:
        print(f"Ошибка сервера ({status}), попробуйте позже")
    else:
        print(f"HTTP ошибка {status}")
        if e.response is not None and e.response.text:
            print(e.response.text[:200])

except requests.exceptions.RequestException as e:
    print(f"Не удалось создать пост: {e}")
    
```
