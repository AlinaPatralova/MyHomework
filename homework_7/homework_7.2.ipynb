## Задание 2. Работа с параметрами запроса

Используйте API OpenWeather для получения данных о погоде.
Напишите программу, которая:

1. Принимает название города от пользователя,
2. Отправляет GET-запрос к API и выводит текущую температуру и описание погоды.

```python
import requests

def get_weather(api_key: str, city: str):
    url = "https://api.openweathermap.org/data/2.5/weather"
    params = {"q": city, "appid": api_key, "units": "metric", "lang": "ru"}
    r = requests.get(url, params=params, timeout=10)
    r.raise_for_status()
    data = r.json()
    temp = data["main"]["temp"]
    desc = data["weather"][0]["description"]
    name = data.get("name", city)
    return f"Сейчас в {name}: {temp:.1f}°C, {desc}"

if __name__ == "__main__":
    api_key = input("Вставь свой OpenWeather API key: ").strip()
    city = input("Город: ").strip()
    try:
        print(get_weather(api_key, city))
    except requests.exceptions.HTTPError as e:
        if e.response is not None and e.response.status_code == 404:
            print("Город не найден")
        elif e.response is not None and e.response.status_code == 401:
            print("Неверный ключ API")
        else:
            print("HTTP-ошибка:", e)
    except requests.exceptions.RequestException as e:
        print("Сетевая ошибка:", e)
```
