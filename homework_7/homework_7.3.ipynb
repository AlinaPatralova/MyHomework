## Задание 3. Создание и обработка POST-запросов

Выберите API, который позволяет создавать ресурсы. Например, JSONPlaceholder.
Напишите программу, которая:

1. Отправляет POST-запрос для создания нового поста,
2. Выводит ID созданного поста и его содержимое.

```python
import json
import requests

api_url = "https://jsonplaceholder.typicode.com/posts"

new_post = {
    "title": "Привет, POST!",
    "body": "Тестовый контент"
}

try:
    r = requests.post(api_url, json=new_post, timeout=10)
    r.raise_for_status()
    data = r.json()

    post_id = data.get("id")
    print(f"ID созданного поста: {post_id}")
    print("Содержимое созданного поста:")
    print(json.dumps(data, ensure_ascii=False))

except requests.exceptions.RequestException as e:
    print(f"Не удалось создать пост: {e}")
```
